```python
import numpy as np
categories = ["Code", "Inline Comments/Docstrings", "Tests"]
x = np.array([1,2,3])
pyxem = [10581, 11540, 13169]
py4dstem = [38582, 16295, 519]
libertem = [25323, 8598, 20930]
```

```python
plt.style.use('dark_background') #lets use dark mode just for fun :)\
fig, axs = plt.subplots(1)
axs.bar(x,pyxem, width=.3, label="pyxem", color="lightcoral")
axs.bar(x+.3,py4dstem, width=0.3, label="py4dstem",color="skyblue")
axs.bar(x+.6,libertem, width=0.3, label="libertem",color="darkslategray")
axs.set_xticks(x+0.3)
axs.set_xticklabels(categories)
axs.legend()
axs.set_ylabel("Number of Lines of Code")
fig.savefig("CodeComparison.png", dpi=500)
```
